### **Data Preparation - Data Cleaning**

Checks for missing values, duplicates

```python
# Check for missing values
df.isnull().sum()
```

```python
# Check for missing values (percentage)
df.isnull().mean()*100
```

```python
# Check for duplicates
duplicates = df[df.duplicated()]

# Print duplicates (if any)
if not duplicates.empty:
    print("Duplicate rows found:")
    print(duplicates)
else:
    print("No duplicates found.")
```

```python
# Replace missing values in 'director' column with 'Unknown'
df['director'] = df['director'].fillna('Unknown')

# Replace missing values in 'cast' column with 'Unknown'
df['cast'] = df['cast'].fillna('Unknown')

# Replace missing values in 'cast' column with 'Unknown'
df['country'] = df['country'].fillna('Unknown')
```

Given that the percentage of missing values for *director*, *cast*, and *country* variables is more than 5%, removing null value would affect the data. Using a custom placeholder that indicates missingness 'Unknown' for type objects would be more suitable.

```python
# Convert the 'date_added' column to datetime, handling errors and inconsistencies
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y', errors='coerce')

# Step 1: Calculate the mean date (ignoring NaT values)
mean_date = df['date_added'].mean()

# Step 2: Replace NaT (null) values with the mean date
df['date_added'] = df['date_added'].fillna(mean_date)
```

```python
# Drop null values for 'rating' and 'duration'
df.dropna(subset=['rating', 'duration'], inplace=True)
```

Given that the percentage of missing values for *rating* and *duration* variables is less than 5%, removing null value wouldn't affect the data.
Removing null values would be more suitable.

```python
# Convert 'date_added' to datetime format
df['date_added'] = pd.to_datetime(df['date_added'])

# Create 'month_added' and 'year_added' columns
df['month_added'] = df['date_added'].dt.month
df['year_added'] = df['date_added'].dt.year
```

```python
# Convert 'duration' column to numeric for TV Shows and Movies
# Extract numeric part, handle seasons/episodes for TV Shows
# Check if the 'duration' column is of type object (string) before applying str.extract
if df['duration'].dtype == object:
    df['duration'] = pd.to_numeric(df['duration'].str.extract('(\d+)')[0], errors='coerce')

# For TV Shows, set duration to number of seasons
# Similarly, check if the 'duration' column is of type object before applying str.extract
if df.loc[df['type'] == 'TV Show', 'duration'].dtype == object:
    df.loc[df['type'] == 'TV Show', 'duration'] = pd.to_numeric(df.loc[df['type'] == 'TV Show', 'duration'].str.extract('(\d+)')[0], errors='coerce')

# Calculate the correlation matrix only for numeric columns
correlation_matrix = df.select_dtypes(include=np.number).corr()

# Set up the figure size
plt.figure(figsize=(10, 8))

# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Set title and show the plot
plt.title("Correlation Heatmap for Netflix Data", fontsize=16)
plt.show()
```

- Very weak positive correlation (0.11) between release_year and year_added - this suggests the slight tendency for newer releases to be added in more recent years
- No correlation between month_added & duration (0.01) and year_added and duration (0.01) - this suggests that these two variables are not related
- Weak negative correlation between release_year and duration (-0.25) - this suggests that as the release year increases, the duration of the content slights decreases, or vice versa

### **Percentage of Movies and TV Shows**

1. What is the breakdown of content on Netflix in terms of movies versus TV shows?

```python
# Count the frequency of each unique value in 'type' column
count = df['type'].value_counts()

# Calculate the percentage of each unique value
percent = df['type'].value_counts(normalize=True) * 100

# Create the frequency table
Freq_table = pd.DataFrame({"Frequency": count, "Percentage": percent})

# Plot a pie chart with both number (frequency) and percentage in the labels
plt.figure(figsize=(8, 6))

# Custom function to format the autopct labels with both frequency and percentage
def func(pct, allvals):
    absolute = int(pct / 100.*sum(allvals))  # Absolute value
    return f"{absolute}\n({pct:.1f}%)"  # Show both number and percentage

plt.pie(count, labels=count.index, autopct=lambda pct: func(pct, count), startangle=90, colors=['lightcoral', 'lightgreen', 'lightblue'])
plt.title("Distribution of 'type' Values")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

*   Movies make up 69.6% of the dataset, with a total of 6125 entries
*   TV Shows make up 30.4% of the dataset, with a total of 2674 entries

Analysis: Netflix's primary content emphasis is on movies, which suggests that there is a higher demand for movies from its audience

### **Countries Producing Content**

4. What are the top five countries producing Netflix’s movies and TV shows?

```python
# Step 1: Filter data for Movies and TV Shows
df_movies = df[df['type'] == 'Movie']
df_tv_shows = df[df['type'] == 'TV Show']

# Step 2: Filter out rows where 'country' is "Unknown" for both Movies and TV Shows
df_movies = df_movies[df_movies['country'] != 'Unknown']
df_tv_shows = df_tv_shows[df_tv_shows['country'] != 'Unknown']

# Step 3: Count the frequency of each unique value in 'country' for Movies
count_movies = df_movies['country'].value_counts()

# Step 4: Count the frequency of each unique value in 'country' for TV Shows
count_tv_shows = df_tv_shows['country'].value_counts()

# Step 5: Create frequency tables for Movies and TV Shows
freq_table_movies = pd.DataFrame({"Frequency": count_movies}).reset_index()
freq_table_movies = freq_table_movies.rename(columns={"index": "country"})
freq_table_movies = freq_table_movies.sort_values(by='Frequency', ascending=False).head(5)

freq_table_tv_shows = pd.DataFrame({"Frequency": count_tv_shows}).reset_index()
freq_table_tv_shows = freq_table_tv_shows.rename(columns={"index": "country"})
freq_table_tv_shows = freq_table_tv_shows.sort_values(by='Frequency', ascending=False).head(5)

# Step 6: Plot the bar chart for Movies and TV Shows in two subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot for Movies
axes[0].barh(freq_table_movies['country'], freq_table_movies['Frequency'], color='skyblue', edgecolor='black')
axes[0].set_title('Top 5 Countries by Frequency (Movies)', fontsize=14)
axes[0].set_xlabel('Frequency', fontsize=12)
axes[0].set_ylabel('Country', fontsize=12)

# Add frequency as text annotation for Movies
for bar in axes[0].patches:
    width = bar.get_width()
    axes[0].text(width + 5, bar.get_y() + bar.get_height() / 2, f'{int(width)}', va='center', ha='left', color='black')

# Plot for TV Shows
axes[1].barh(freq_table_tv_shows['country'], freq_table_tv_shows['Frequency'], color='lightgreen', edgecolor='black')
axes[1].set_title('Top 5 Countries by Frequency (TV Shows)', fontsize=14)
axes[1].set_xlabel('Frequency', fontsize=12)
axes[1].set_ylabel('Country', fontsize=12)

# Add frequency as text annotation for TV Shows
for bar in axes[1].patches:
    width = bar.get_width()
    axes[1].text(width + 5, bar.get_y() + bar.get_height() / 2, f'{int(width)}', va='center', ha='left', color='black')

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

- The top three countries movies are produced in are 1. United States, 2. India, 3. United Kingdom, 4. Canada, and 5. Spain
- The top three countries TV Shows are produced in are 1. United States, 2. United Kingdom, 3. Japan, 4. South Korea, and 5. India

Analysis: United States dominates the production of movies and TV shows in Netflix. While United Kingdom and India both ranks as significant contributors, India and United Kingdom have higher counts in movies and TV shows respectively. Emerging markets in Spain, Japan, and South Korea signals a shift towards more diverse interional content

### **Highest Release Year**

8. Which year had the highest number of movies and TV shows released on Netflix?

```python
# Filter dataset for Movies and TV Shows
movie = df[df['type'] == 'Movie']
TVshow = df[df['type'] == 'TV Show']

# Get value counts of 'year_added' column for Movies and TV Shows
movie_year_counts = movie['year_added'].value_counts().sort_index()
TVshow_year_counts = TVshow['year_added'].value_counts().sort_index()

# Combine both dataframes into one for plotting
years_df = pd.DataFrame({
    'year_added': movie_year_counts.index.tolist() + TVshow_year_counts.index.tolist(),
    'count': movie_year_counts.values.tolist() + TVshow_year_counts.values.tolist(),
    'type': ['Movie']*len(movie_year_counts) + ['TV Show']*len(TVshow_year_counts)
})

# Plot Bar Chart
plt.figure(figsize=(14, 7))
ax = sns.barplot(x='year_added', y='count', hue='type', data=years_df, palette='Set2')

# Add titles and labels
plt.title('Number of Movies and TV Shows Released by Year', fontsize=16)
plt.xlabel('Year Added', fontsize=12)
plt.ylabel('Number of Releases', fontsize=12)
plt.xticks(rotation=45)  # Rotate the x-axis labels for better readability
plt.grid(True)

# Add value labels on top of each bar
for p in ax.patches:
    # Get the height (value) of the bar
    height = p.get_height()
    # Get the position of the bar (x and y coordinates)
    x = p.get_x() + p.get_width() / 2
    y = p.get_height()

    # Add the count value above the bar
    ax.text(x, y + 10, f'{int(height)}', ha='center', va='bottom', fontsize=10, color='black')

# Show the plot
plt.show()
```

- 2017 - 2019 shows consistent increase in both Movies and TV Shows, where the number released more than quadruple (from 251 to 1,424 and from 164 to 673 respectively)
- 2019 indicates the highest number of Movies (1,424) and TV Shows (673) releases
- 2020 - 2021 shows a slight decrease for both Movies and TV Shows as compared to 2019

Analysis:
- Pre-pandemic growth (2016-2019) showed a strong upward trend in both Movies and TV Shows, likely due to the increasing popularity of streaming platforms and global content production.
- COVID-19 disruptions (2020-2021) slowed movie releases significantly, but TV shows adapted better due to faster production cycles and the ability to release content digitally.

### **Top Release Months**

- Top three months for movie releases are July, December, January/October.
- Top three months for TV shows releases are May, July, and December.

Analysis:
- Summer and Winter are prime periods for movies and TV shows releases, with July seeing the highest number of launches.
- This indicates that both the summer break and the winter season are key windows for rolling out new shows.
